# üåä Interpolasi Polinomial Spline untuk Kedalaman Laut üåä

Proyek ini berfokus pada penggunaan **interpolasi polinomial spline** untuk memperkirakan kedalaman laut antara Teluk Suak dan Pulau Lemukutan. Kami menerapkan interpolasi spline dengan derajat yang berbeda, yaitu derajat 1 (linier), derajat 2 (kuadratik), derajat 3 (kubik), dan derajat 4 (kuartik). Tujuan utama dari analisis ini adalah untuk membandingkan kinerja masing-masing derajat spline dalam merepresentasikan data kedalaman laut berdasarkan dataset yang diberikan.

Dataset ini diambil dari jurnal penelitian:
> Hariski, Muhammad & Kusumastuti, Nilamsari & Yudhi, Yudhi. (2024). CUBIC SPLINE INTERPOLATION TO APPROXIMATE SEA DEPTH BETWEEN TELUK SUAK AND LEMUKUTAN ISLAND. EPSILON: JURNAL MATEMATIKA MURNI DAN TERAPAN (EPSILON: JOURNAL OF PURE AND APPLIED MATHEMATICS). 18. 1. 10.20527/epsilon.v18i1.10813.  
> [Link ke Jurnal](https://www.researchgate.net/publication/384757784_CUBIC_SPLINE_INTERPOLATION_TO_APPROXIMATE_SEA_DEPTH_BETWEEN_TELUK_SUAK_AND_LEMUKUTAN_ISLAND)

## üìä Dataset

Data yang digunakan terdiri dari titik-titik kedalaman laut (meter) pada berbagai jarak (kilometer) antara Teluk Suak dan Pulau Lemukutan. Data ini menjadi dasar untuk eksperimen interpolasi dalam studi kasus ini.

### Contoh Data:

| Jarak (km) | Kedalaman Laut (m) |
|------------|---------------------|
| 0          | 0.00               |
| 0.325      | -0.02              |
| 3.7        | -7.59              |
| 15.9       | 0.00               |
| ...        | ...                |

## üìê Metode Interpolasi

Dalam kode ini, digunakan beberapa metode interpolasi polinomial spline untuk mendekati nilai kedalaman laut di antara titik data yang diketahui:

1. **Spline Linier (Derajat 1)**: Menghasilkan fungsi linear antar titik data.
2. **Spline Kuadratik (Derajat 2)**: Menggunakan polinomial derajat 2 untuk memperkirakan kurva antar titik data.
3. **Spline Kubik (Derajat 3)**: Menggunakan polinomial derajat 3 untuk menghasilkan kurva yang lebih halus, seperti yang disarankan dalam jurnal ini.
4. **Spline Kuartik (Derajat 4)**: Menggunakan polinomial derajat 4 untuk pendekatan yang lebih kompleks.

## ‚öôÔ∏è Fitur Kode

- **Penghitungan Error**: Menghitung MSE (Mean Squared Error), RMSE (Root Mean Squared Error), MAE (Mean Absolute Error), dan MAPE (Mean Absolute Percentage Error) untuk setiap derajat interpolasi. Hal ini bertujuan untuk mengevaluasi seberapa baik masing-masing metode interpolasi mendekati data aktual.
- **Penggunaan Titik Antara**: Menggunakan interpolasi linier untuk mengestimasi nilai kedalaman laut pada titik-titik antara untuk analisis yang lebih mendalam terhadap performa interpolasi.
- **Visualisasi Grafis**: Menyajikan plot interpolasi dari masing-masing derajat untuk memvisualisasikan seberapa baik kurva interpolasi dalam mendekati data kedalaman laut.

## üöÄ Hasil & Pembahasan

### Persamaan Interpolasi

Berikut adalah beberapa persamaan interpolasi yang dihasilkan oleh kode:

- **Spline Linier**: Persamaan berupa fungsi linier antara dua titik data. Misalnya:
  \[
  f(x) = -0.061538 \cdot x + 0.000000 \quad \text{untuk interval } [0, 0.325]
  \]

- **Spline Kubik**: Menghasilkan persamaan polinomial kubik untuk setiap interval. Contoh:
  \[
  f(x) = -0.117245 \cdot (x - 0)^3 + 0.158518 \cdot (x - 0)^2 - 0.100673 \cdot (x - 0) + 0.000000 \quad \text{untuk interval } [0, 0.325]
  \]

> Untuk derajat 2 (kuadratik) dan derajat 4 (kuartik), persamaan eksplisit antar interval tidak disajikan oleh library yang digunakan.

### Evaluasi Error

Dari perhitungan MAPE pada titik-titik antara, berikut adalah ringkasan hasil error dari masing-masing derajat interpolasi:

| Derajat | MSE     | RMSE     | MAE      | MAPE   |
|---------|---------|----------|----------|--------|
| 1       | 0.0000  | 0.0000   | 0.0000   | 0.00%  |
| 2       | 0.0462  | 0.2149   | 0.1439   | 2.87%  |
| 3       | 0.0515  | 0.2270   | 0.1513   | 3.02%  |
| 4       | 1.0796  | 1.0390   | 0.7895   | 22.15% |

Dari tabel di atas, dapat disimpulkan bahwa:

- **Spline Linier (Derajat 1)**: Memberikan hasil MAPE yang sangat rendah (0.00%), yang berarti interpolasi ini sangat tepat dalam kasus dataset ini.
- **Spline Kubik (Derajat 3)**: Seperti yang diusulkan dalam jurnal, spline kubik memberikan hasil yang cukup baik dengan MAPE sekitar 3.02%.
- **Spline Kuartik (Derajat 4)**: Meskipun menghasilkan MAPE lebih tinggi, spline kuartik memberikan pendekatan yang lebih kompleks. Namun, dalam kasus ini, derajat 4 tidak menghasilkan hasil yang lebih baik dibanding derajat yang lebih rendah.

### Visualisasi

Visualisasi interpolasi untuk setiap derajat spline menunjukkan perbedaan dalam kedalaman interpolasi di setiap interval, di mana spline kubik terlihat memberikan kurva yang lebih halus sesuai dengan data kedalaman laut.

## üîç Kesimpulan

Dari hasil eksperimen ini, spline linier dan kubik menunjukkan performa yang baik untuk dataset ini. Spline kubik, yang menjadi fokus dalam jurnal, memberikan hasil yang akurat dan dapat merepresentasikan perubahan kedalaman laut dengan cukup baik. Meskipun spline kuartik menawarkan pendekatan yang lebih kompleks, hasil MAPE yang tinggi menunjukkan bahwa pendekatan ini tidak selalu cocok untuk semua kasus, khususnya ketika data menunjukkan perubahan yang cukup sederhana.

```python
import numpy as np
import pandas as pd
from scipy.interpolate import interp1d, CubicSpline, UnivariateSpline
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt

# Data Kedalaman Laut Antara Teluk Suak dan Pulau Lemukutan
data_x = np.array([0, 0.325, 0.812, 1.06, 1.2, 1.48, 2, 3, 3.7, 4,
                   5, 5.5, 5.83, 6, 7, 8.19, 9, 9.5, 10.4, 11,
                   11.5, 12, 12.5, 12.9, 13.2, 14, 15, 15.5, 15.6, 15.7, 15.9])
data_y = np.array([0, -0.02, -0.04, -0.06, -0.07, -0.17, -0.86, -4.19, -7.59, -8.7,
                   -13.4, -15.2, -15.62, -15.33, -11.33, -8.13, -9.11, -10.71, -13.61, -12.67,
                   -13.97, -16.97, -22.07, -23.67, -23.26, -16.4, -5.2, -0.8, -0.4, -0.2, 0])

# Pilih derajat spline (1 hingga 4)
degrees = [1, 2, 3, 4]
results_tables = {}
interpolated_error_metrics = {}

# Fungsi untuk menghitung metrik error, mengabaikan nilai nol untuk MAPE
def calculate_error_metrics(y_actual, y_pred):
    mse = mean_squared_error(y_actual, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_actual, y_pred)

    # Hanya menghitung MAPE untuk nilai y_actual yang tidak nol
    mask = y_actual != 0
    if np.any(mask):  # Pastikan ada nilai yang tidak nol
        mape = np.mean(np.abs((y_actual[mask] - y_pred[mask]) / y_actual[mask])) * 100
    else:
        mape = np.nan  # Jika tidak ada nilai yang valid untuk MAPE
    return mse, rmse, mae, mape

# Titik-titik antara (intermediate points) untuk perhitungan MAPE yang lebih representatif
x_intermediate = np.linspace(min(data_x), max(data_x), 50)
y_actual_intermediate = np.interp(x_intermediate, data_x, data_y)  # Estimasi "aktual" untuk titik antara

# Iterasi untuk setiap derajat
for degree in degrees:
    if degree == 1:
        # Spline linier
        spline = interp1d(data_x, data_y, kind='linear', fill_value="extrapolate")
        print("\n--- Persamaan Spline Linier ---")
        for i in range(len(data_x) - 1):
            slope = (data_y[i+1] - data_y[i]) / (data_x[i+1] - data_x[i])
            intercept = data_y[i] - slope * data_x[i]
            print(f"Interval [{data_x[i]}, {data_x[i+1]}]: f(x) = {slope:.6f} * x + {intercept:.6f}")

    elif degree == 2:
        # Spline kuadratik
        spline = interp1d(data_x, data_y, kind='quadratic', fill_value="extrapolate")
        print("\n--- Persamaan Spline Kuadratik ---")
        print("Tidak tersedia persamaan eksplisit untuk setiap interval pada spline kuadratik.")

    elif degree == 3:
        # Spline kubik
        spline = CubicSpline(data_x, data_y)
        print("\n--- Persamaan Spline Kubik ---")
        for i in range(len(data_x) - 1):
            a, b, c, d = spline.c[:, i]
            x0 = data_x[i]
            print(f"Interval [{data_x[i]}, {data_x[i+1]}]:")
            print(f"f(x) = {a:.6f} * (x - {x0})^3 + {b:.6f} * (x - {x0})^2 + {c:.6f} * (x - {x0}) + {d:.6f}")

    elif degree == 4:
        # Spline kuartik (derajat 4)
        spline = UnivariateSpline(data_x, data_y, k=4)
        print("\n--- Persamaan Spline Kuartik ---")
        print("UnivariateSpline tidak menyediakan koefisien interval secara langsung untuk derajat 4.")
        print("Namun, hasil interpolasi dan galat dihitung dan divisualisasikan.")

    # Prediksi nilai pada titik data asli untuk menghitung galat
    y_pred = spline(data_x)
    galat = np.abs(data_y - y_pred)

    # Buat tabel hasil untuk titik data asli
    results_table = pd.DataFrame({
        "X": data_x,
        "Y Aktual": data_y,
        "Y Prediksi": y_pred,
        "Galat/Error": galat
    }).round(6)

    # Perhitungan dan tabel untuk titik-titik antara
    y_pred_intermediate = spline(x_intermediate)
    mse, rmse, mae, mape = calculate_error_metrics(y_actual_intermediate, y_pred_intermediate)
    interpolated_error_metrics[degree] = (mse, rmse, mae, mape)

    # Buat tabel hasil untuk titik-titik antara
    intermediate_table = pd.DataFrame({
        "X Intermediate": x_intermediate,
        "Y Aktual (Interpolated)": y_actual_intermediate,
        "Y Prediksi)": y_pred_intermediate,
        "Galat/Error)": np.abs(y_actual_intermediate - y_pred_intermediate)
    }).round(6)

    # Simpan tabel untuk derajat tertentu
    results_tables[degree] = (results_table, intermediate_table)

    # Cetak tabel dan metrik error untuk setiap derajat
    print(f"\n=== Hasil untuk Derajat {degree} ===")
    print("\nData Asli:")
    print(results_table)
    print("\nData Titik Antara Menggunakan Spline Linier untuk Y Aktual:")
    print(intermediate_table)
    print("\nEvaluasi Error (berdasarkan titik antara):")
    print(f"  MSE  : {mse:.6f}")
    print(f"  RMSE : {rmse:.6f}")
    print(f"  MAE  : {mae:.6f}")
    print(f"  MAPE : {mape:.2f}%")
    print("=" * 50)

# Plot hasil interpolasi
plt.figure(figsize=(12, 8))
plt.plot(data_x, data_y, 'o', label="Data Asli", markersize=8, color="blue")

# Titik interpolasi untuk kurva yang lebih halus
x_interp = np.linspace(min(data_x), max(data_x), 100)

# Warna untuk tiap derajat
colors = {1: 'orange', 2: 'green', 3: 'purple', 4: 'brown'}

# Plot untuk tiap derajat
for degree in degrees:
    if degree == 1:
        spline = interp1d(data_x, data_y, kind='linear', fill_value="extrapolate")
        y_interp = spline(x_interp)
    elif degree == 2:
        spline = interp1d(data_x, data_y, kind='quadratic', fill_value="extrapolate")
        y_interp = spline(x_interp)
    elif degree == 3:
        spline = CubicSpline(data_x, data_y)
        y_interp = spline(x_interp)
    elif degree == 4:
        spline = UnivariateSpline(data_x, data_y, k=4)
        y_interp = spline(x_interp)

    # Plot hasil interpolasi
    plt.plot(x_interp, y_interp, label=f"Interpolasi Derajat {degree}", color=colors[degree])

plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.minorticks_on()
plt.grid(which='minor', linestyle=':', linewidth=0.5)
plt.xlabel("Jarak (km)")
plt.ylabel("Kedalaman Laut (m)")
plt.legend()
plt.title("Interpolasi Spline dengan Berbagai Derajat")
plt.show()
```
