# üìä Sea Depth Approximation Using Polynomial Spline Interpolation

This repository implements various polynomial spline interpolation techniques to approximate sea depth data between Teluk Suak and Pulau Lemukutan. The code explores linear, quadratic, cubic, and quartic splines, providing visualizations, error metrics, and interval-based polynomial equations for each spline degree.

## üèûÔ∏è Dataset

The dataset used for this project represents the sea depth profile between **Teluk Suak** and **Pulau Lemukutan**. The depth data is approximated at various distance intervals. This study is based on the following journal:

**Reference**:  
Hariski, M., Kusumastuti, N., & Yudhi, Y. (2024). *Cubic Spline Interpolation to Approximate Sea Depth Between Teluk Suak and Lemukutan Island*. ## üíª Features

- **Linear, Quadratic, Cubic, and Quartic Splines**: Supports multiple degrees of spline interpolation.
- **Error Metrics**: Calculates Mean Squared Error (MSE), Root Mean Squared Error (RMSE), Mean Absolute Error (MAE), and Mean Absolute Percentage Error (MAPE).
- **Interval-based Polynomial Equations**: Displays the piecewise equations for each interval for linear and cubic splines.
- **Visualizations**: Graphical comparison of the actual data with interpolated curves for each spline degree.

## üßÆ Methods

This project uses four types of polynomial spline interpolation methods to approximate the sea depth data. Here‚Äôs a detailed description of each approach:

### 1. Linear Spline
- A **piecewise linear approximation** connecting each data point.
- The linear spline is the simplest form of interpolation where each segment between two points is a straight line.
- Provides interval-based linear equations, which are displayed in the output for detailed inspection.
  
### 2. Quadratic Spline
- Fits **quadratic polynomials** between data points.
- It offers more flexibility than linear splines and provides a smoother curve, though it doesn't ensure a smooth first derivative between intervals.
- Due to the complexity of representing each interval as a quadratic equation, explicit equations for each segment are not provided.

### 3. Cubic Spline
- The **most commonly used spline** type due to its smoothness.
- Ensures continuity in the **first and second derivatives** across intervals, providing a very smooth curve that best fits most datasets.
- Provides interval-based cubic polynomial equations in the output, allowing detailed inspection of each interval.

### 4. Quartic Spline
- A **higher-order spline** that aims to reduce error by adding more degrees of freedom.
- Uses SciPy's `UnivariateSpline` with a degree of 4, providing more flexibility.
- Explicit interval equations are not provided due to the complex nature of higher-order polynomials.

## üìä Results and Discussion

Each spline method's effectiveness was evaluated using the following metrics:

- **MSE (Mean Squared Error)**: Represents the average squared difference between predicted and actual values.
- **RMSE (Root Mean Squared Error)**: The square root of MSE, providing a measure of error in the same units as the original data.
- **MAE (Mean Absolute Error)**: The average absolute difference between the predicted and actual values.
- **MAPE (Mean Absolute Percentage Error)**: A percentage representation of the error, excluding points where the actual value is zero to avoid division errors.

The table below summarizes the performance of each spline degree:

| Spline Degree | MSE     | RMSE   | MAE    | MAPE    |
|---------------|---------|--------|--------|---------|
| Linear        | 0.0000  | 0.0000 | 0.0000 | 0.00%   |
| Quadratic     | 0.0462  | 0.2149 | 0.1439 | 2.87%   |
| Cubic         | 0.0515  | 0.2270 | 0.1513 | 3.02%   |
| Quartic       | 1.0796  | 1.0390 | 0.7895 | 22.15%  |

### üîç Analysis

- **Linear Spline**:
  - Captures the general shape of the dataset but results in sharp, angular transitions between points due to its piecewise nature.
  - Produces the lowest error metrics but sacrifices smoothness.
  
- **Quadratic Spline**:
  - Offers a more flexible fit than linear, resulting in lower error metrics compared to cubic and quartic.
  - However, it does not handle sharp changes effectively as it lacks the smooth continuity offered by cubic splines.

- **Cubic Spline**:
  - **The best-performing spline** in terms of both smoothness and accuracy.
  - Ensures smooth transitions across intervals, balancing low error and continuity, making it the preferred method for this dataset.

- **Quartic Spline**:
  - Adds more flexibility with higher degrees but can lead to **overfitting**, as indicated by higher error values.
  - The high MAPE value suggests that the model struggled with generalization, and additional degrees of freedom may have introduced unnecessary complexity.

**‚ú® Summary**:  
This project uses polynomial spline interpolation to approximate sea depth between Teluk Suak and Pulau Lemukutan. By comparing four spline types, we found that **cubic splines** offered the best combination of smoothness and accuracy for this dataset. The linear spline showed accurate, albeit sharp, predictions, while the quartic spline overfitted the data, leading to higher errors. The detailed plots, equations, and error metrics provide a comprehensive understanding of the strengths and weaknesses of each interpolation method.

```python
import numpy as np
import pandas as pd
from scipy.interpolate import interp1d, CubicSpline, UnivariateSpline
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt

# Data Kedalaman Laut Antara Teluk Suak dan Pulau Lemukutan
data_x = np.array([0, 0.325, 0.812, 1.06, 1.2, 1.48, 2, 3, 3.7, 4,
                   5, 5.5, 5.83, 6, 7, 8.19, 9, 9.5, 10.4, 11,
                   11.5, 12, 12.5, 12.9, 13.2, 14, 15, 15.5, 15.6, 15.7, 15.9])
data_y = np.array([0, -0.02, -0.04, -0.06, -0.07, -0.17, -0.86, -4.19, -7.59, -8.7,
                   -13.4, -15.2, -15.62, -15.33, -11.33, -8.13, -9.11, -10.71, -13.61, -12.67,
                   -13.97, -16.97, -22.07, -23.67, -23.26, -16.4, -5.2, -0.8, -0.4, -0.2, 0])

# Pilih derajat spline (1 hingga 4)
degrees = [1, 2, 3, 4]
results_tables = {}
interpolated_error_metrics = {}

# Fungsi untuk menghitung metrik error, dengan pengaturan tambahan pada MAPE
def calculate_error_metrics(y_actual, y_pred, epsilon=1e-5):
    mse = mean_squared_error(y_actual, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_actual, y_pred)

    # Menambahkan epsilon ke y_actual untuk menghindari pembagian dengan nol atau angka terlalu kecil
    y_actual_safe = np.where(np.abs(y_actual) < epsilon, epsilon, y_actual)

    # Hitung MAPE hanya untuk nilai yang tidak berlebihan (tidak melebihi 1000%)
    mape_values = np.abs((y_actual - y_pred) / y_actual_safe) * 100
    mape_values = np.where(mape_values > 1000, 0, mape_values)  # Ignorasi nilai MAPE yang terlalu besar

    # Rata-rata MAPE setelah mengabaikan nilai yang tidak masuk akal
    mape = np.mean(mape_values)

    return mse, rmse, mae, mape

# Titik-titik antara (intermediate points) untuk perhitungan MAPE yang lebih representatif
x_intermediate = np.linspace(min(data_x), max(data_x), 50)
y_actual_intermediate = np.interp(x_intermediate, data_x, data_y)  # Estimasi "aktual" untuk titik antara

# Iterasi untuk setiap derajat
for degree in degrees:
    if degree == 1:
        # Spline linier
        spline = interp1d(data_x, data_y, kind='linear', fill_value="extrapolate")
        print("\n--- Persamaan Spline Linier ---")
        for i in range(len(data_x) - 1):
            slope = (data_y[i+1] - data_y[i]) / (data_x[i+1] - data_x[i])
            intercept = data_y[i] - slope * data_x[i]
            print(f"Interval [{data_x[i]}, {data_x[i+1]}]: f(x) = {slope:.6f} * x + {intercept:.6f}")

    elif degree == 2:
        # Spline kuadratik
        spline = interp1d(data_x, data_y, kind='quadratic', fill_value="extrapolate")
        print("\n--- Persamaan Spline Kuadratik ---")
        print("Tidak tersedia persamaan eksplisit untuk setiap interval pada spline kuadratik.")

    elif degree == 3:
        # Spline kubik
        spline = CubicSpline(data_x, data_y)
        print("\n--- Persamaan Spline Kubik ---")
        for i in range(len(data_x) - 1):
            a, b, c, d = spline.c[:, i]
            x0 = data_x[i]
            print(f"Interval [{data_x[i]}, {data_x[i+1]}]:")
            print(f"f(x) = {a:.6f} * (x - {x0})^3 + {b:.6f} * (x - {x0})^2 + {c:.6f} * (x - {x0}) + {d:.6f}")

    elif degree == 4:
        # Spline kuartik (derajat 4)
        spline = UnivariateSpline(data_x, data_y, k=4)
        print("\n--- Persamaan Spline Kuartik ---")
        print("UnivariateSpline tidak menyediakan koefisien interval secara langsung untuk derajat 4.")
        print("Namun, hasil interpolasi dan galat dihitung dan divisualisasikan.")

    # Prediksi nilai pada titik data asli untuk menghitung galat
    y_pred = spline(data_x)
    galat = np.abs(data_y - y_pred)

    # Buat tabel hasil untuk titik data asli
    results_table = pd.DataFrame({
        "X": data_x,
        "Y Aktual": data_y,
        "Y Prediksi": y_pred,
        "Galat/Error": galat
    }).round(6)

    # Perhitungan dan tabel untuk titik-titik antara
    y_pred_intermediate = spline(x_intermediate)
    mse, rmse, mae, mape = calculate_error_metrics(y_actual_intermediate, y_pred_intermediate)
    interpolated_error_metrics[degree] = (mse, rmse, mae, mape)

    # Buat tabel hasil untuk titik-titik antara
    intermediate_table = pd.DataFrame({
        "X Intermediate": x_intermediate,
        "Y Aktual (Interpolated)": y_actual_intermediate,
        "Y Prediksi)": y_pred_intermediate,
        "Galat/Error)": np.abs(y_actual_intermediate - y_pred_intermediate)
    }).round(6)

    # Simpan tabel untuk derajat tertentu
    results_tables[degree] = (results_table, intermediate_table)

    # Cetak tabel dan metrik error untuk setiap derajat
    print(f"\n=== Hasil untuk Derajat {degree} ===")
    print("\nData Asli:")
    print(results_table)
    print("\nData Titik Antara Menggunakan Spline Linier untuk Y Aktual:")
    print(intermediate_table)
    print("\nEvaluasi Error (berdasarkan titik antara):")
    print(f"  MSE  : {mse:.6f}")
    print(f"  RMSE : {rmse:.6f}")
    print(f"  MAE  : {mae:.6f}")
    print(f"  MAPE : {mape:.2f}%")
    print("=" * 50)

# Plot hasil interpolasi dan titik ujinya
plt.figure(figsize=(14, 8))
# Plot data asli dengan tanda bulat
plt.plot(data_x, data_y, 'o', label="Original Data Points", markersize=8, color="blue", markeredgecolor='black')
# Plot titik uji (titik antara) dengan tanda silang
plt.plot(x_intermediate, y_actual_intermediate, 'x', label="Test Points (Intermediate)", markersize=6, color="red", markeredgewidth=1)
# Titik interpolasi untuk kurva yang lebih halus
x_interp = np.linspace(min(data_x), max(data_x), 100)
# Warna untuk tiap derajat spline
colors = {1: 'orange', 2: 'green', 3: 'purple', 4: 'brown'}

# Plot untuk tiap derajat spline
for degree in degrees:
    if degree == 1:
        spline = interp1d(data_x, data_y, kind='linear', fill_value="extrapolate")
        y_interp = spline(x_interp)
        linestyle = '--'  # Dotted line for linear
    elif degree == 2:
        spline = interp1d(data_x, data_y, kind='quadratic', fill_value="extrapolate")
        y_interp = spline(x_interp)
        linestyle = '-.'  # Dash-dot line for quadratic
    elif degree == 3:
        spline = CubicSpline(data_x, data_y)
        y_interp = spline(x_interp)
        linestyle = '-'  # Solid line for cubic
    elif degree == 4:
        spline = UnivariateSpline(data_x, data_y, k=4)
        y_interp = spline(x_interp)
        linestyle = ':'  # Dotted line for quartic

    # Plot hasil interpolasi
    plt.plot(x_interp, y_interp, label=f"Spline Interpolation (Degree {degree})", color=colors[degree], linestyle=linestyle, linewidth=2)

# Enhanced grid and labels
plt.grid(True, which='both', linestyle='--', linewidth=0.6, color='grey')
plt.minorticks_on()
plt.grid(which='minor', linestyle=':', linewidth=0.4, color='lightgrey')
plt.xlabel("Distance (km)", fontsize=14, fontweight='bold')
plt.ylabel("Sea Depth (m)", fontsize=14, fontweight='bold')
plt.title("Sea Depth Spline Interpolation with Different Degrees and Test Points", fontsize=16, fontweight='bold')
plt.legend(loc="best", fontsize=10, frameon=True, shadow=True, fancybox=True, borderpad=1)
plt.tight_layout()
plt.show()
```
